```python
import pandas as pd
import datetime as dt
```

```python
temps = pd.read_csv('Content_work_order.csv')
temps = temps.dropna(how='all')
del temps['field_wo_supervisor_uid']
del temps['field_wo_job_type_value']
del temps['filed_wo_job_class_value']
del temps['field_wo_report_date_value']
del temps['filed_wo_zone_nid']
del temps['field_wo_center_cost_value']
del temps['filed_woparent_nid']
del temps['field_wo_team_nid']
temps = temps.dropna(subset=['field_wo_execution_date_value', 'field_wo_finish_date_value'])

```

```python
temps.reset_index(drop=True, inplace=True)
```
